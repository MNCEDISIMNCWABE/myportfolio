```python
def null_checker(df):
    '''
    Returns the count of rows found in the data frame where there is a null present.

    Parameters:
        df (str): Pandas dataframe

    Returns:
        Prints the null count, null percentage, and row count per column.
    '''
    row_count = len(df)
    null_counts = df.isnull().sum().sort_values(ascending=False)
    null_percentages = (null_counts / row_count) * 100
    combined = pd.concat([null_counts, null_percentages], axis=1, keys=['Null Count', 'Null Percentage'])
    combined['Row Count'] = row_count
    print(combined)

print('test_target:')
null_checker(test_target)
print('\ntest_demographics:')
null_checker(test_demographics)
print('\ntest_credit_file:')
null_checker(test_credit_file)
print('\ntest_ob_transactions:')
null_checker(test_ob_transactions)
```

```python
def preprocess_data(df):
    '''
    Drops columns with more than 70% of missing data and excludes fields that are unnecessary for the analysis.

    Parameters:
        df (str): Pandas dataframe

    Returns:
        returns a dataframe with not fields missing >70% of data
    '''
    
    # Drop features with >70% missing data if they exist
    missing_threshold = 0.7
    columns_to_drop = [col for col in df.columns if (df[col].isnull().mean() > missing_threshold)]
    df = df.drop(columns=columns_to_drop, axis=1)

    # Select features for analysis based on existing columns in each dataframe 
    selected_columns = ['user_id', 'ageOfCreditFileInMonths', 'ageOfAccountInMonths',
                        'balanceAmount', 'creditLimitAmount', 'paymentFrequency',
                        'statusOne', 'statusThree', 'statusThreePlus', 'statusTwo', 'statusZeroOrBetter',
                        'utilisation','amount', 'categoryConfidence', 'balance',
                       'credit_score','salary','employment_status', 'accountType','onePlusInWindow',
                        'twoPlusInWindow','defaultInWindow']

    selected_columns = [col for col in selected_columns if col in df.columns]
    df = df[selected_columns]

    return df

# Call the function for dataframes
test_credit_file = preprocess_data(test_credit_file)
test_ob_transactions = preprocess_data(test_ob_transactions)
test_demographics = preprocess_data(test_demographics)
test_target = preprocess_data(test_target)
```

```python
def handle_missing_values(df):
    '''
    Imputes missing values for numerical and categorical features.

    Parameters:
        df (str): Pandas dataframe

    '''
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col].fillna(df[col].mean(), inplace=True)
        else:
            df[col].fillna(df[col].mode()[0], inplace=True)
    
    return df

# Handle missing values for test_demographics
test_demographics = handle_missing_values(test_demographics)

# Handle missing values for test_ob_transactions
test_ob_transactions = handle_missing_values(test_ob_transactions)

# Handle missing values for test_credit_file
test_credit_file = handle_missing_values(test_credit_file)
```

```python
def aggregate_data(df, groupby_cols, agg_funcs):
   
    '''
    Aggregates the numerical fields per user_id

    Parameters:
        df (str): Pandas dataframe

    '''

    aggregated_df = df.groupby(groupby_cols).agg(agg_funcs).reset_index()
    return aggregated_df

# Define aggregation functions for each dataframe
ob_transactions_agg_funcs = {
    'amount': 'mean',
    'categoryConfidence': 'mean',
    'balance': 'mean'
}

demographics_agg_funcs = {
    'credit_score': 'mean',
    'salary': 'mean',
}

credit_file_agg_funcs = {
    'ageOfCreditFileInMonths': 'max',
    'ageOfAccountInMonths':'max',
    'balanceAmount':'mean',
    'creditLimitAmount':'mean',
    'paymentFrequency':'first',
    'statusOne':'max',
    'statusTwo':'max',
    'statusThree':'max',
    'statusThreePlus':'max',
    'statusZeroOrBetter':'max',
    'utilisation':'mean'
}

# Call the function for each dataframe
test_ob_transactions = aggregate_data(test_ob_transactions, ['user_id'], ob_transactions_agg_funcs)
test_demographics = aggregate_data(test_demographics, ['user_id', 'employment_status'], demographics_agg_funcs)
test_credit_file = aggregate_data(test_credit_file, ['user_id','accountType'], credit_file_agg_funcs)
```

```python
def join_dataframes(target_df, demographics_df, credit_file_df, ob_transactions_df):
    '''
    Combines all dataframes to one dataframe based on user_id

    Parameters:
        df (str): Pandas dataframe
    '''
    # Join the dataframes on 'user_id'
    joined_df = target_df.merge(demographics_df, on='user_id', how='inner')
    joined_df = joined_df.merge(credit_file_df, on='user_id', how='inner')
    joined_df = joined_df.merge(ob_transactions_df, on='user_id', how='inner')
    
    return joined_df

# Call the function to join dataframes
joined_df = join_dataframes(test_target, test_demographics, test_credit_file, test_ob_transactions)
```

```python
def one_hot_key_encoding(df):
    '''
    One hotkey encode column.
            Parameters:
                    df (str): Pandas dataframe
                    column_name (str): Column name
            Returns:
                    Dataframe with one hotkey encoded column
    '''

    one_hot_employment_status_category = pd.get_dummies(df.employment_status, prefix='employment_status',dtype=int)
    df = df.drop('employment_status',axis = 1)
    df = df.join(one_hot_employment_status_category)
    one_hot_accountType_category = pd.get_dummies(df.accountType,prefix = 'accountType',dtype=int)
    df = df.drop('accountType',axis = 1)
    df = df.join(one_hot_accountType_category)
    one_hot_paymentFrequency_category = pd.get_dummies(df.paymentFrequency,prefix = 'paymentFrequency',dtype=int)
    df = df.drop('paymentFrequency',axis = 1)
    df = df.join(one_hot_paymentFrequency_category)
    return df

joined_df = one_hot_key_encoding(joined_df)
```

```python
def calculate_predictive_power(data_frame, target_column):
    '''
    Calculates predictive power of independent features on target feature.
            Parameters:
                    df (str): Pandas dataframe
                    target_column (int): target_column
            Returns:
                    Dataframe with showing predictive power score (pps) for each independent feature.
    '''
    data_frame[target_column] = data_frame[target_column].astype(str)
    predictors_df = pps.predictors(data_frame, target_column)
    predictors_df = predictors_df.sort_values(by="ppscore", ascending=False)
    return predictors_df

# call function
predictors_result = calculate_predictive_power(joined_df, "defaultInWindow")
predictors_result
```

```python
def set_features_and_target(df):
    '''
    Returns two data frames with features and target variables.
            Parameters:
                    df (str): Pandas dataframe
    '''
    X = df.drop(['defaultInWindow'], axis=1)
    y = df['defaultInWindow']
    
    return X,y
                 
X,y = set_features_and_target(joined_df)               
```

```python
def train_test_split_df(X,y):
    '''
    Create train test split.
            Parameters:
                    df (str): Pandas dataframe
            Returns:
                    Train and test dataframes
    '''
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = train_test_split_df(X,y)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
def model_application(X_train,y_train,optimiser):
    # Building model
    '''
    Model application. If optimiser is true , a grid search is applied to optimise the model. 
    If optimiser is false base model is used
            Parameters:
                    X_train (str): Pandas dataframe of feature set training variables
                    y_train (str): Pandas dataframe of predictor variable for training set.
                    optimiser (str): Boolean to indicate optimisation required. 
            Returns:
                    Random Forest model.
    '''
    if optimiser == True:
                # Building our model with K-fold validation and GridSearch to find the best parameters

                # Defining all the parameters
                params = {
                'max_features': [1, 3, 10],
                'min_samples_split': [2, 3, 10],
                'min_samples_leaf': [1, 3, 10],
                'criterion': ["entropy", "gini"]
                }

                cv = StratifiedShuffleSplit(n_splits=5, test_size=0.3, random_state=15)

                # Building model
                rf = RandomForestClassifier()

                # Parameter estimating using GridSearch
                grid = GridSearchCV(rf, param_grid=params, scoring='accuracy', n_jobs =-1, cv=cv, verbose=1)
                
                # Fitting the model
                grid.fit(X_train, y_train)

                dtc_grid_val_score = grid.best_score_
                print('Best Score:', dtc_grid_val_score)
                print('Best Params:', grid.best_params_)
                print('Best Estimator:', grid.best_estimator_)

                rf_clf = grid.best_estimator_
    else:
                rf_clf = RandomForestClassifier(n_estimators=150,
                                                max_depth=16,
                                                criterion='gini',
                                                class_weight='balanced',
                                                min_samples_split=10,
                                                max_leaf_nodes=20,
                                                max_features=0.5,
                                                bootstrap=True)
                rf_clf.fit(X_train, y_train)

    return rf_clf

optimiser = False
rf_clf = model_application(X_train,y_train,optimiser)
```

```python
def model_predict(X_test):
    '''
    Create y_pred , model prediction based on test set features.
            Parameters:
                    X_test (str): Pandas dataframe
            Returns:
                    y_pred
    '''
    y_pred = rf_clf.predict(X_test)
    
    return y_pred

y_pred = model_predict(X_test)
print('Training Accuracy Score(%):',rf_clf.score(X_train,y_train)*100)
print('Test Accuracy Score(%):',rf_clf.score(X_test,y_test)*100)
```

```python
def generate_confusion_matrix(y_test,y_pred):
    '''
    Generates confusion matrix.
            Parameters:
                    y_test (array): numpy array of actual values
                    y_pred (array): numpy array of predicted values
            Returns:
                    Y_pred
    '''
    
    cm = confusion_matrix(y_test, y_pred, labels=rf_clf.classes_)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=rf_clf.classes_)
    disp.plot()
    plt.show()
    
    return cm

generate_confusion_matrix(y_test,y_pred)
```

```python
def generate_perfomance_metrics(y_test,y_pred):
    '''
    Generates perfomance metrics precision, recall, f1 score, accuracy
            Parameters:
                    y_test (array): numpy array of actual values
                    y_pred (array): numpy array of predicted values
            Returns:
                    y_pred
    '''
    rf_model_score = accuracy_score(y_test, y_pred)
    print('Model Accuracy:', rf_model_score)
    return print('Classification Report:\n', classification_report(y_test, y_pred))

generate_perfomance_metrics(y_test,y_pred)
```

```python
def plot_feature_importance(model, feature_names, n=40):
    
    '''
    Produces a plot displaying features that are most important predictors of the target feature.
            Parameters:
                    random forest model
                    feature names
    '''
    
    plt.figure(figsize=(12, 10))

    feat_importances = pd.Series(model.feature_importances_, index=feature_names)
    feat_importances.nsmallest(n).plot(kind='barh')
    importances = model.feature_importances_
    indices = np.argsort(importances)

    plt.xlabel('Importance')
    plt.ylabel('Features')
    plt.title("Feature Importance - Ranked", fontsize=15)
    plt.show()

plot_feature_importance(rf_clf, X_train.columns)
```
