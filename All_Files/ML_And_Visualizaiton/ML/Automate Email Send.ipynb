```python
import smtplib, ssl
from email.mime.base import MIMEBase
from email import encoders
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
import os
from datetime import datetime

def send_mail_with_excel(subject):
    today = datetime.now().strftime("%Y-%m-%d")
    EMAIL_ADDRESS = os.environ.get('EMAIL_USER')
    EMAIL_PASSWORD = os.environ.get('EMAIL_PASSWORD')

    if subject == 'OOS Historic and Futuristic':
        # recipients = ['keletso.dithebe@takealot.com','byran.titus@takealot.com', 'ceril.mhaule@takealot.com', 'flint.ntshangase@takealot.com', 'bulungisa.jarana@takealot.com', 'maria.louw@takealot.com', 'asisipho.gqoli@takealot.com', 'liezl.bam@takealot.com', 'zamasomi.Msomi@takealot.com', 'direshan.pillay@takealot.com','desigan.moodley@takealot.com']
        test_recipients = ['keletso.dithebe@takealot.com','byran.titus@takealot.com', 'mncedisi.mncwabe@takealot.com', 'ceril.mhaule@takealot.com']
        msg = MIMEMultipart("alternative")
        msg['Subject'] = subject
        msg['From'] = EMAIL_ADDRESS
        msg['To'] = ", ".join(test_recipients)
        html = """\
        <html>
        <body>
            <p>Good day,<br>
            Please find attached OOS Reports.<br>
            Also the Historic report and Futuristic report can now be found <a href="https://docs.google.com/spreadsheets/d/1xr9HoTEv7QlxwLTMg2JrlWEkcCo8htifHQ-aY_vqkEQ/edit#gid=0">here</a> <br>
            Kind regards<br>
            Merchant Analytics<br>
            </p>
        </body>
        </html>
        """
        part1 = MIMEText(html, "html")
        msg.attach(part1)

        with open('FuturisticView_'+today+'.xlsx', "rb") as attachment:
        # Add file as application/octet-stream
        # Email client can usually download this automatically as attachment
            part2 = MIMEBase("application", "octet-stream")
            part2.set_payload(attachment.read())

        # Encode file in ASCII characters to send by email    
        encoders.encode_base64(part2)
        part2.add_header(
            "Content-Disposition",
            f"attachment; filename= {'FuturisticView_'+today+'.xlsx'}",
        )
        msg.attach(part2)
        # text = msg.as_string()

        with open('HistoricView_'+today+'.xlsx', "rb") as attachment:
        # Add file as application/octet-stream
        # Email client can usually download this automatically as attachment
            part3 = MIMEBase("application", "octet-stream")
            part3.set_payload(attachment.read())

        # Encode file in ASCII characters to send by email    
        encoders.encode_base64(part3)
        part3.add_header(
            "Content-Disposition",
            f"attachment; filename= {'HistoricView_'+today+'.xlsx'}",
        )
        msg.attach(part3)

        
        text = msg.as_string()
        context = ssl.create_default_context()
        with smtplib.SMTP_SSL('smtp.gmail.com', 465, context=context) as smtp:
            smtp.login(EMAIL_ADDRESS, EMAIL_PASSWORD)
            smtp.sendmail(EMAIL_ADDRESS, test_recipients, text)

    elif subject == 'Replen OOS Report':
        # recipients = ['keletso.dithebe@takealot.com','byran.titus@takealot.com', 'ceril.mhaule@takealot.com', 'flint.ntshangase@takealot.com', 'bulungisa.jarana@takealot.com', 'maria.louw@takealot.com', 'asisipho.gqoli@takealot.com', 'liezl.bam@takealot.com', 'zamasomi.Msomi@takealot.com', 'direshan.pillay@takealot.com','desigan.moodley@takealot.com']
        test_recipients = ['keletso.dithebe@takealot.com', 'byran.titus@takealot.com', 'mncedisi.mncwabe@takealot.com', 'ceril.mhaule@takealot.com']
        msg = MIMEMultipart("alternative")
        msg['Subject'] = subject
        msg['From'] = EMAIL_ADDRESS
        msg['To'] = ", ".join(test_recipients)
        html = """\
        <html>
        <body>
            <p>Good day,<br>
            Please find attached Replen OOS Reports.<br>
            Also the Replen OOS report can now be found <a href="https://docs.google.com/spreadsheets/d/1xr9HoTEv7QlxwLTMg2JrlWEkcCo8htifHQ-aY_vqkEQ/edit#gid=66449188">here</a> <br>
            Kind regards<br>
            Merchant Analytics<br>
            </p>
        </body>
        </html>
        """
        part1 = MIMEText(html, "html")
        msg.attach(part1)

        with open('Replen_HistoricView_'+today+'.xlsx', "rb") as attachment:
        # Add file as application/octet-stream
        # Email client can usually download this automatically as attachment
            part2 = MIMEBase("application", "octet-stream")
            part2.set_payload(attachment.read())

        # Encode file in ASCII characters to send by email    
        encoders.encode_base64(part2)
        part2.add_header(
            "Content-Disposition",
            f"attachment; filename= {'Replen_HistoricView_'+today+'.xlsx'}",
        )
        msg.attach(part2)
        text = msg.as_string()
        context = ssl.create_default_context()
        with smtplib.SMTP_SSL('smtp.gmail.com', 465, context=context) as smtp:
            smtp.login(EMAIL_ADDRESS, EMAIL_PASSWORD)
            smtp.sendmail(EMAIL_ADDRESS, test_recipients, text)

    return 
```
