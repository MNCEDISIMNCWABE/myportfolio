```python
import smtplib, ssl
from email.mime.base import MIMEBase
from email import encoders
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
import os
from datetime import datetime

def send_mail_with_excel(subject):
    today = datetime.now().strftime("%Y-%m-%d")
    EMAIL_ADDRESS = os.environ.get('EMAIL_USER')
    EMAIL_PASSWORD = os.environ.get('EMAIL_PASSWORD')

    if subject == 'OOS Historic and Futuristic':
        
        test_recipients = ['mncedisi.mncwabe@takealot.com']
        msg = MIMEMultipart("alternative")
        msg['Subject'] = subject
        msg['From'] = EMAIL_ADDRESS
        msg['To'] = ", ".join(test_recipients)
        html = """\
        <html>
        <body>
            <p>Good day,<br>
            Please find attached report.<br>
            it can be found via this link<a href="https://docs.google.com/spreadsheets/d/1HcBdIlDqc8PQoATRtnaFFWWd4gGT7AeY-O0bqKN_3DQ/edit#gid=716408807">here</a> <br>
            Kind regards<br>
            Unathi<br>
            </p>
        </body>
        </html>
        """
        part1 = MIMEText(html, "html")
        msg.attach(part1)

        with open('Future.xlsx', "rb") as attachment:
        # Add file as application/octet-stream
        # Email client can usually download this automatically as attachment
            part2 = MIMEBase("application", "octet-stream")
            part2.set_payload(attachment.read())

        # Encode file in ASCII characters to send by email    
        encoders.encode_base64(part2)
        part2.add_header(
            "Content-Disposition",
            f"attachment; filename= {'Future.xlsx'}",
        )
        msg.attach(part2)
        # text = msg.as_string()

        with open('History.xlsx', "rb") as attachment:
        # Add file as application/octet-stream
        # Email client can usually download this automatically as attachment
            part3 = MIMEBase("application", "octet-stream")
            part3.set_payload(attachment.read())

        # Encode file in ASCII characters to send by email    
        encoders.encode_base64(part3)
        part3.add_header(
            "Content-Disposition",
            f"attachment; filename= {'History.xlsx'}",
        )
        msg.attach(part3)

        
        text = msg.as_string()
        context = ssl.create_default_context()
        with smtplib.SMTP_SSL('smtp.gmail.com', 465, context=context) as smtp:
            smtp.login(EMAIL_ADDRESS, EMAIL_PASSWORD)
            smtp.sendmail(EMAIL_ADDRESS, test_recipients, text)

    elif subject == 'Replen OOS Report':
        test_recipients = ['mncedisi.mncwabe@takealot.com']
        msg = MIMEMultipart("alternative")
        msg['Subject'] = subject
        msg['From'] = EMAIL_ADDRESS
        msg['To'] = ", ".join(test_recipients)
        html = """\
        <html>
        <body>
            <p>Good day,<br>
            Please find attached report.<br>
            It can be found via this link <a href="https://docs.google.com/spreadsheets/d/1HcBdIlDqc8PQoATRtnaFFWWd4gGT7AeY-O0bqKN_3DQ/edit#gid=716408807">here</a> <br>
            Kind regards<br>
            Unathi<br>
            </p>
        </body>
        </html>
        """
        part1 = MIMEText(html, "html")
        msg.attach(part1)

        with open('Replen.xlsx', "rb") as attachment:
        # Add file as application/octet-stream
        # Email client can usually download this automatically as attachment
            part2 = MIMEBase("application", "octet-stream")
            part2.set_payload(attachment.read())

        # Encode file in ASCII characters to send by email    
        encoders.encode_base64(part2)
        part2.add_header(
            "Content-Disposition",
            f"attachment; filename= {'Replen.xlsx'}",
        )
        msg.attach(part2)
        text = msg.as_string()
        context = ssl.create_default_context()
        with smtplib.SMTP_SSL('smtp.gmail.com', 465, context=context) as smtp:
            smtp.login(EMAIL_ADDRESS, EMAIL_PASSWORD)
            smtp.sendmail(EMAIL_ADDRESS, test_recipients, text)

    return
```
