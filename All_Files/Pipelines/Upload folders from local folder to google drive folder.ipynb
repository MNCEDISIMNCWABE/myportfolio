```python
from google.oauth2 import service_account
from googleapiclient.discovery import build
import os

# Define the scopes and folder paths
SCOPES = ['https://www.googleapis.com/auth/drive']
SERVICE_ACCOUNT_FILE = 'C:/Users/leemn/OneDrive/Documents/Personal Google Service Account/bright-arc-328707-b5e2d782b48b.json'  
FOLDER_PATHS = [
    "C:/Users/leemn/OneDrive/Documents/Crypto", 
    "C:/Users/leemn/OneDrive/Documents/Competitions", 
    "C:/Users/leemn/OneDrive/Documents/Deployments",
    "C:/Users/leemn/OneDrive/Documents/Pipelines",
    "C:/Users/leemn/OneDrive/Documents/Route_Optimization",
    "C:/Users/leemn/OneDrive/Documents/ML_And_Visualizaiton"
]
PARENT_FOLDER_ID = '1tDftQP1FG5SECRU6Z71GbSAn2YTxgwa_'  # ID of the target Google Drive folder

def authenticate():
    credentials = service_account.Credentials.from_service_account_file(
        SERVICE_ACCOUNT_FILE, scopes=SCOPES)
    return credentials

def upload_file(service, file_name, parent_id):
    file_metadata = {'name': os.path.basename(file_name), 'parents': [parent_id]}
    media = MediaIoBaseUpload(io.BytesIO(open(file_name, "rb").read()), mimetype='application/octet-stream')
    file = service.files().create(body=file_metadata, media_body=media, fields='id').execute()
    print(f'Uploaded {file_name} with file ID {file["id"]}')

def upload_folder(service, folder_path, parent_id):
    folder_name = os.path.basename(folder_path)
    file_metadata = {'name': folder_name, 'mimeType': 'application/vnd.google-apps.folder', 'parents': [parent_id]}
    folder = service.files().create(body=file_metadata, fields='id').execute()
    new_parent_id = folder.get('id')
    for item in os.listdir(folder_path):
        item_path = os.path.join(folder_path, item)
        if os.path.isfile(item_path):
            upload_file(service, item_path, new_parent_id)
        elif os.path.isdir(item_path):
            upload_folder(service, item_path, new_parent_id)

def main():
    creds = authenticate()
    service = build('drive', 'v3', credentials=creds)
    for path in FOLDER_PATHS:
        upload_folder(service, path, PARENT_FOLDER_ID)

if __name__ == '__main__':
    main()
```
