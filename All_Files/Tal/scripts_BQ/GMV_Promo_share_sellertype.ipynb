```python

SELECT
bh.bh_division AS Division,
bh.bh_reporting_department AS Department,
FORMAT_DATE("%Y-%m",order_placed_timestamp) AS YearMonth,
CASE 
   WHEN oi.is_marketplace_item = false THEN "Retail"
   ELSE "Marketplace"
END AS SellerType,
CASE 
   WHEN oi.is_deal=true THEN "Promo"
   ELSE "Non-Promo"
END AS isPromo,
round(SUM(oi.order_item_sales),2) AS SalesEx
FROM `tal-production-data-bi.tal_dm_order.orders_2010_2020` od,
UNNEST(od.order_items) oi
INNER JOIN `tal-production-data-bi.tal_dm_product.business_hierarchy` bh
  ON bh.product_id = oi.product_id
WHERE   DATE(order_authed_timestamp) >= "2018-04-01" AND  DATE(order_authed_timestamp) < "2020-01-01" 
        AND DATE(order_placed_timestamp) >= "2018-04-01" AND DATE(order_placed_timestamp) < "2020-01-01"
        AND oi.is_authed_not_return_cancel IS TRUE 
        AND oi.order_item_status NOT IN ('Return Canceled','Canceled')
        AND oi.order_item_status LIKE 'Shipped'
        AND order_authorisation_status IN ('Auth')
        AND is_returned = false
        AND is_cancelled = false
        AND is_voucher = false
        AND is_test_customer = false
GROUP BY YearMonth, SellerType, isPromo, bh.bh_division, bh.bh_reporting_department
UNION ALL
SELECT
bh.bh_division AS Division,
bh.bh_reporting_department AS Department,
FORMAT_DATE("%Y-%m",order_placed_timestamp) AS YearMonth,
CASE 
   WHEN oi.is_marketplace_item = false THEN "Retail"
   ELSE "Marketplace"
END AS SellerType,
CASE 
   WHEN oi.is_deal=true THEN "Promo"
   ELSE "Non-Promo"
END AS isPromo,
round(SUM(oi.order_item_sales),2) AS SalesEx
FROM `tal-production-data-bi.tal_dm_order.orders_2020_2030` od,
UNNEST(od.order_items) oi
INNER JOIN `tal-production-data-bi.tal_dm_product.business_hierarchy` bh
  ON bh.product_id = oi.product_id
WHERE   DATE(order_authed_timestamp) >= "2020-01-01" AND  DATE(order_authed_timestamp) < "2022-04-01" 
        AND DATE(order_placed_timestamp) >= "2020-01-01" AND DATE(order_placed_timestamp) < "2022-04-01"
        AND oi.is_authed_not_return_cancel IS TRUE 
        AND oi.order_item_status NOT IN ('Return Canceled','Canceled')
        AND oi.order_item_status LIKE 'Shipped'
        AND order_authorisation_status IN ('Auth')
        AND is_returned = false
        AND is_cancelled = false
        AND is_voucher = false
        AND is_test_customer = false
GROUP BY YearMonth, SellerType, isPromo, bh.bh_division, bh.bh_reporting_department
ORDER BY YearMonth ASC
```
